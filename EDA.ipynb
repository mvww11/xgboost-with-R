# Análise Exploratória de Dados
## Nota de Matemática: a variável que queremos prever

Primeiramente, vamos conhecer nossa variável dependente (aquela que queremos prever).

Temos uma variável com assimetria positiva. Com uma média de 523 pontos, temos 75% dos candidatos pontuando até 598. Os outros 25% dos candidatos pontuaram até 985. Menos de 1% dos candidatos obtiveram mais que 811 pontos.

```r
#histograma de notas
ggplot(y, aes(x=NU_NOTA_MT)) + geom_histogram()
```

## Features
Vamos explorar as variáveis (features) que temos disponíveis para prever a nota de Matemática de cada estudante.

Nossas features são todas categóricas, com exceção de "quantas pessoas moram na sua residência?" e "idade".

### Renda da Família

O gráfico em forma de escada mostra que à medida que estudantes sobem nas classes de renda, também sobem as médias das notas obtidas por eles. A renda parece ter forte influencia no desempenho do ENEM e será uma variável importante na regressão da nota.

Vemos que 46,7% dos alunos da maior classe de renda conseguem uma nota superior a 700 pontos, necessária para ingressar nos cursos mais concorridos, enquanto que apenas 1,2% dos estudantes de menor renda realizam o mesmo feito.

Mesmo para baixa renda, existem alunos com ótimo desempenho, mas são outliers (esses são os casos de sucesso que vemos na TV, de alunos humildes que superaram as dificuldades e conseguiram ótimo resultado). São raros, contudo. Somente 0.003% dos alunos da menor classe de renda conseguiram chegar a 800 pontos no ENEM (esse número passa para 13% na maior classe de renda).

```r
#boxplot de renda
ggplot(data=full_data, aes(x=Q006, y=NU_NOTA_MT)) + geom_boxplot()
```

### Quantas pessoas moram na sua residência

Estudantes que moram em residências com até 4 pessoas têm nota pouco maior que aqueles que moram em residências mais populosas. Mas essa variável não parece ser tão significativa na regressão.

Ideia para feature engineering: a partir da razão entre renda e moradores na residência, podemos criar a "renda per capita" do candidato, que, intuitivamente, parece ser uma variável importante para a regressão.

```r
#boxplot de número de residentes
ggplot(data=full_data, aes(x=Q005, y=NU_NOTA_MT)) + geom_boxplot()
```

### Sexo
Existe uma diferença significativa entre as médias das notas de homens em relação às mulheres, sendo a média dos primeiros superior em cerca de 30 pontos.

```r
#boxplot de sexo
ggplot(data=full_data, aes(x=TP_SEXO, y=NU_NOTA_MT)) + geom_boxplot()
```

### Escolaridade Máxima

A escolaridade máxima entre pai e mãe aponta ganhos suaves na nota à medida que o aluno avança nas primeiras 4 classes. Esse ganho é mais acentuado da 4ª para a 5ª, e da 5ª para a 6ª classes, que correspondem ao caso em que algum dos pais possui escolaridade em nível superior.

```r
#boxplot de escolaridade máxima
ggplot(data=full_data, aes(x=max_escol, y=NU_NOTA_MT)) + geom_boxplot()
```

### Renda per capita
A renda per capita apresenta formato de escada até a 12ª classe de renda, a partir da qual fica relativamente estável.

```r
#boxplot de renda per capita
ggplot(data=full_data, aes(x=perCapita, y=NU_NOTA_MT)) + geom_boxplot()
```
